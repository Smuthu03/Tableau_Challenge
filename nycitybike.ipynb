## Data Clensing of City-Bike data to use the dataset for Tableau Visualization 

```python

# File to Load (Remember to Change These)
city_latlon_file = Path("city_latlon.csv")

# # Read School and Student Data File and store into Pandas DataFrames
city_latlon_df = pd.read_csv(city_latlon_file)

city_latlon_df
```

```python
from geopy.extra.rate_limiter import RateLimiter
from geopy.geocoders import Nominatim
import pandas as pd
geolocator = Nominatim(user_agent="user_agent")
# add rate limit
reverse = RateLimiter(geolocator.reverse, min_delay_seconds=1)

# create dataframe
df = pd.DataFrame({"geo_loc"  :[(40.71, -74.04 ) ,(40.72, -74.05) ,(40.72, -74.04) ,(40.71, -74.03) ,(40.73, -74.07) ,
                                (40.71, -74.05) ,(40.72, -74.06) ,(40.72, -74.03) , (40.71, -74.07) ,
                                (40.73, -74.05) ,(40.73, -74.03) ,(40.72, -74.03)  ,(40.72, -74.03) ,(40.74, -74.07) ,
                                (40.73, -74.04) ,(40.71, -74.07) ,(40.72, -74.07) , (40.701, -74.04 ) ,(40.74, -74.03) ,
                                (40.72, -74.05) ,(40.73, -74.06) ,( 40.72, -74.06) ,(40.72, -74.04) ,
                                (40.74, -74.04) ,(40.74, -74.08) , (40.71, -74.06) ],
                   "latitude_2" : [40.71 , 40.72 ,40.72 ,40.71 ,40.73, 40.71 ,40.72 ,40.72 , 40.71 ,40.73 ,40.73 ,
                                 40.72 ,40.72 ,40.74,40.73,40.71 ,40.72 , 40.701 ,40.74 ,40.72,40.73 , 40.72,
                                 40.72,40.74 ,40.74 , 40.71],
                  "longitude_2" : [ -74.04 , -74.05, -74.04, -74.03, -74.07, -74.05, -74.06, -74.03, -74.07,
                                   -74.05, -74.03, -74.03, -74.03, -74.07, -74.04, -74.07, -74.07, -74.04, -74.03, 
                                   -74.05, -74.06, -74.06, -74.04 ,-74.04, -74.08, -74.06]})
# get location coordinates
geo_loc  = df.geo_loc.values
```

```python
state_list = [] # Initialize empty dict
city_list = [] # Initialize empty dict
county_list =[]
zip_code =[]
for cord in geo_loc:
    # send request
    location = reverse(cord, addressdetails=True)
    # get state value
    state = location.raw["address"].get("state")
    city = location.raw["address"].get("city")
    county = location.raw["address"].get("county")
    zip = location.raw["address"].get("postcode")
    # store state value
    state_list.append(state)
    city_list.append(city)
    county_list.append(county)
    zip_code.append(zip)
#assign back states
df['states'] = state_list
df['city'] = city_list
df['county'] = county_list
df['zip_code'] = zip_code
# print(geo_loc)
df
```

```python
city_latlon_df1 = city_latlon_df.merge(df, left_on=['latitude_2', 'longitude_2'], right_on=['latitude_2', 'longitude_2'])
city_latlon_df1
```
